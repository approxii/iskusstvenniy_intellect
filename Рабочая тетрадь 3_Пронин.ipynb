```python
#1.3.1

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def printdistance(x, y):
  print("Для", x, "и", y)
  print("Расстояние Евклида", np.linalg.norm(np.array(x) - np.array(y)))
  print("Квадрат Евклидова расстояния", np.linalg.norm(np.array(x) - np.array(y)) ** 2)
  print("Расстояние Чебышева", np.linalg.norm(np.array(x) - np.array(y), ord=np.inf))
  print("Расстояние Хемминга", np.linalg.norm(np.array(x) - np.array(y), ord=1))
  fig = plt.figure()
  ax = fig.add_subplot(111, projection = '3d')
  ax.scatter(*zip(x))
  ax.scatter(*zip(y))
  plt.show()
  print("\n")

a = (0, 0, 0)
b = (3, 3, 3)
c = (6, 7, 2)
d = (8, 3, 6)

printdistance(a, b)
printdistance(a, c)
printdistance(a, d)
printdistance(b, c)
printdistance(b, d)
printdistance(c, d)
```

```python
#2.3.1

from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

K = 3
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

print( '(-2, -2) is class')
print(model.predict([[-2, -2]]))

print('(1, 3) is class')
print(model.predict([[1, 3]]))

K = 1
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

print( '(-2, -2) is class')
print(model.predict([[-2, -2]]))

print('(1, 3) is class')
print(model.predict([[1, 3]]))

K = 5
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

print( '(-2, -2) is class')
print(model.predict([[-2, -2]]))

print('(1, 3) is class')
print(model.predict([[1, 3]]))


import matplotlib.pyplot as plt
from operator import irshift
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

iris = sns.load_dataset('iris')
print(iris)
print("\n")

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15
)


print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print("\n")

print(x_train.head())
print("\n")

print(y_train.head())
print("\n")

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)


plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data = iris, hue='species', s=70)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend (loc=2)
plt.grid()

for i in range(len(y_test)):
  if np.array(y_test)[i] != y_pred[i]:
    plt.scatter(x_test.iloc[i, 3], x_test.iloc[1, 2], color="red", s=150)

from sklearn.metrics import accuracy_score 
print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')
```
